# 探索测试集

```python
test_data.shape
```

# 合并训练集和测试集

```python
all_data = pd.concat((train_data, test_data)).reset_index(drop=True)
all_data.shape
```

```python
# 缺失值统计
null_value = all_data.isna().sum().sort_values(ascending=False)
null_value[null_value > 0]
```

# 缺失值处理

```python
# 剔除缺失值最多的 PoolQC、MiscFeature、Alley、Fence、FireplaceQu 特征，以及 Id 特征
all_data.drop(['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu', 'Id'], axis=1, inplace=True)
null_value = all_data.isna().sum().sort_values(ascending=False)
null_value[null_value > 0]
```

```python
# 将含有缺失值的特征名添加到列表，并剔除 预测特征
null_list = []
for i in null_value[null_value > 0].keys():
    null_list.append(i)
null_list.remove('SalePrice')
null_list
```

```python
# 将缺失值列表又拆分为 分类特征、数值特征
null_cat = []
null_num = []
for col in null_list:
    if all_data[col].dtype in [object, bool] and len(all_data[col].unique()) <= 50:
        null_cat.append(col)
    else:
        null_num.append(col)
null_cat, null_num
```

```python
# 对缺失值中的分类特征使用众数填充缺失值
for col in null_cat:
    all_data[col] = all_data[col].fillna(max(list(all_data[col]), key=list(all_data[col]).count))
    
# 数值特征使用中位数填充缺失值
for col in null_num:
    all_data[col] = all_data[col].fillna(all_data[col].median())    

# 判断缺失值是否填充完成
null_value = all_data.isna().sum().sort_values(ascending=False)
null_value[null_value > 0]
```

# 数据类型转化

```python
# 寻找总数据集中的分类特征
categorical_col = []
for col in all_data.columns:
    if all_data[col].dtype in [object, bool] and len(all_data[col].unique()) <= 50:
        categorical_col.append(col)
categorical_col
```

```python
# 将非数字的标签分类字段编码为数字
for col in categorical_col:
    all_data[col] = all_data[col].astype("category").cat.codes

all_data.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(categories='auto')

# 独热编码分类列
# 不过这里得到的是一个数组，因此后续要将它转换为 dataframe
array_hot_encoded = ohe.fit_transform(all_data[categorical_col]).toarray()

# 将数组转换为dataframe
data_hot_encoded = pd.DataFrame(array_hot_encoded, index=all_data.index)

# 得到其他不需要编码的字段
data_other_cols = all_data.drop(columns=categorical_col)

# 连接两个dataframe
all_data = pd.concat([data_hot_encoded, data_other_cols], axis=1)

all_data.head()
```

```python
# 将所有特征的列名称转换为 str
all_data.columns = all_data.columns.astype(str)
```

# 拆分训练集和测试集

```python
train_data.shape, test_data.shape, all_data.shape
```

```python
train_data = all_data.iloc[:1460, :]
test_data = all_data.iloc[-1459:, :]
train_data.shape, test_data.shape
```

```python
# 删除测试集中的 SalePrice 特征
test_data.drop(['SalePrice'], axis=1, inplace=True)
```

# 清除异常值

```python
from sklearn.neighbors import LocalOutlierFactor
import numpy as np

clf = LocalOutlierFactor(contamination = 0.02)
# 识别异常值:
outliers = clf.fit_predict(train_data)
# 删除异常值，其中 1 表示正常值，-1 表示异常值:
train_data_cleaned = train_data[np.where(outliers == 1, True, False)]
```

```python
x_train = train_data_cleaned.drop('SalePrice', axis = 1)
```

```python
y_train = train_data_cleaned[['SalePrice']]
```

# 模型预测

```python
# 网格搜索用于寻找最优参数
from sklearn.model_selection import GridSearchCV

cv_params = {'n_estimators': [400, 500, 600, 700, 800]}
```

```python
# XGBRegressor 模型
from xgboost import XGBRegressor

other_params = {'learning_rate': 0.1, 'n_estimators': 500, 'max_depth': 5, 'min_child_weight': 1, 'seed': 0,
                'subsample': 0.8, 'colsample_bytree': 0.8, 'gamma': 0, 'reg_alpha': 0, 'reg_lambda': 1}

xgb = XGBRegressor(**other_params)
xgb = GridSearchCV(estimator=xgb, param_grid=cv_params, scoring='r2', cv=5, verbose=1, n_jobs=4)
xgb.fit(x_train, y_train)
xgb_prediction = xgb.predict(test_data)
```

```python
# CatBoostRegressor 模型
from catboost import CatBoostRegressor

cat = CatBoostRegressor()
cat = GridSearchCV(estimator=cat, param_grid=cv_params, scoring='r2', cv=5, verbose=1, n_jobs=4)
cat.fit(x_train, y_train)
cat_prediction = cat.predict(test_data)
```

```python
# StackingRegressor 模型
from sklearn.linear_model import RidgeCV
from sklearn.svm import LinearSVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import StackingRegressor

estimators = [
    ('lr', RidgeCV()),
    ('svr', LinearSVR(random_state=42))
]

stacking = StackingRegressor(
    estimators=estimators,
    final_estimator=RandomForestRegressor(n_estimators=10,
                                          random_state=42)
)

stacking.fit(x_train, y_train)
stacking_prediction = stacking.predict(test_data)
```

```python
# 模型融合
prediction = (0.45 * xgb_prediction) + (0.5 * cat_prediction) + (0.05 * stacking_prediction)
```

# 模型评估

```python
from sklearn.metrics import mean_squared_error 

# 融合模型评估
y_predict = (0.45 * xgb.predict(x_train)) + (0.5 * cat.predict(x_train)) + (0.05 * stacking.predict(x_train))
np.sqrt(mean_squared_error(np.log(y_train), np.log(y_predict)))  
```

```python
# xgb 模型评估
np.sqrt(mean_squared_error(np.log(y_train), np.log(xgb.predict(x_train))))  
```

```python
# cat 模型评估
np.sqrt(mean_squared_error(np.log(y_train), np.log(cat.predict(x_train))))  
```

```python
# stacking 模型评估
np.sqrt(mean_squared_error(np.log(y_train), np.log(stacking.predict(x_train))))  
```
